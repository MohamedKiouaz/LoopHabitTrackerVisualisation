```python
import numpy as np
import pandas as pd
import calmap
```

```python
Checkmarks = pd.read_csv('Checkmarks.csv')
Checkmarks['Date'] = pd.to_datetime(Checkmarks['Date'])
Date = Checkmarks['Date']
```

```python
%matplotlib notebook
for i in range(1, Checkmarks.shape[1]):
    calmap.calendarplot(pd.Series(Checkmarks.iloc[:, i].values, index = Date))
```
